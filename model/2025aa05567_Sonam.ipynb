## Model Performance Observations

### Logistic Regression
Logistic Regression performed well on the dataset with balanced precision and recall. Since several medical features show linear relationships with the target variable, the model was able to capture patterns effectively. It provides stable and interpretable results.

### Decision Tree
Decision Tree achieved moderate performance but slightly lower accuracy compared to ensemble methods. This may be due to overfitting, as a single tree is sensitive to small variations in the training data.

### K-Nearest Neighbors (KNN)
KNN showed the lowest performance among all models. As a distance-based algorithm, it is sensitive to high-dimensional feature space created after encoding categorical variables. This likely reduced its classification effectiveness.

### Naive Bayes
Naive Bayes performed surprisingly well despite its strong independence assumption. The probabilistic nature of the model works effectively when features are relatively independent, making it a strong baseline classifier.

### Random Forest (Best Performing Model)
Random Forest achieved the highest Accuracy, F1-score, and MCC. Being an ensemble of multiple decision trees, it reduces overfitting and captures non-linear feature interactions effectively. It is the most robust and reliable model for this dataset.

### XGBoost
XGBoost also performed strongly and close to Random Forest. As a gradient boosting model, it sequentially improves prediction errors. Slightly lower performance compared to Random Forest may be due to default hyperparameters without tuning.

### Overall Conclusion
Ensemble methods (Random Forest and XGBoost) outperformed individual classifiers, indicating that combining multiple weak learners improves predictive performance for this heart disease classification problem.
